```python
students = ['shabista', 'kiran', 'paul', 'reema', 'zain', 'keerthi', 'zainab']
```

```python
**INTRODUCTION TO LOOPING**
```

```python
FOR Loop
Syntax --> 
for tempvar in mainvar:
    print (tempvar)
```

```python
#Example
for x in students:
    print(x)
```

```python
for a in students:
    print(a)
```

```python
for x in students:
    print(f"Keep up the good work, {students[1].title()}")
```

```python
for x in students:
    print(f"Keep up the good work, {x}")
```

```python
for x in students:
    print(f"Keep up the good work, {x}")
    print(f"I wish you keep up the progress, {x}")
```

```python
for x in students:
    print(f"Keep up the good work, {x}")
    print(f"I wish you keep up the progress, {x}\n") #---> .\n can also be used    --- \n can also come out side the F"string
```

```python
for x in students:
    print(f"Keep up the good work, {x}")
    print(f"I wish you keep up the progress, {x}")
print("Loop ends here") #----> END of LOOP denoted bby indentation!! :)
```
