```python
#Continuation with lists
```

```python
my_students = ['hema', 'ravi','suresh', 'anji', 'iqra', 'aisha', 'naveen', 'peter']
```

```python
print(my_students)
```

```python
#Modify element---->
my_students[2] = 'sagarika'
```

```python
print(my_students)
```

```python
#delete element
del my_students[3]
print(my_students)
```

```python
#Second approach to delete--------*** POP METHOD
**It performs temporary delete, maaintains carbon copy of deleted element.
**By default the element will be excluded from the list

```

```python
x = my_students.pop() #pops last element in the list
print(my_students)
```

```python
print(x) #Pooped value comes here!!
```
