```python
#Introduction to list data type:
    
Definition: A list is an ordered collection of items.
    Declare is square bracket
    Classified as mutable  (Can be altered later)
```

```python
students = ['Sneha', 'Adil', 'danish','Priyanka']  #Elements in the list (the names/values)
```

```python
print(students)
```

```python
#How to verify data type:
type(students)
```

```python
#Introduction to indexing-----> 0,1,2,3..... Like an array
print(students[2]) #Will give third value, because index starts with zero
```

```python
#Improvise ----> Title case 
print(students[2].title())
```

```python
#Improvise ----> Upper case 
print(students[2].upper())
```
