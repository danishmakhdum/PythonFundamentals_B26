```python
name = 'kiran kumar'
#Title Case
print(name.title())
```

```python
#Introduction to F Strings
first_name = "rohit"
last_name ="pota"

#Get full name
full_name = f"{first_name}{last_name}"
print(full_name)
```

```python
#Get full name
full_name = f"{first_name} {last_name}"
print(full_name)
```

```python
#Further Enhancement
print(full_name.title())
```

```python
print(full_name.upper())
```
